```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

# Open the Flipkart webpage
url = "https://www.flipkart.com/"
driver = webdriver.Chrome()  # Make sure you have the ChromeDriver executable in your PATH
driver.get(url)

# Step 2: Enter "sunglasses" in the search field and click the search icon
product = driver.find_element(By.XPATH, '/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input')
product.send_keys("sunglasses")
product.send_keys(Keys.RETURN)

# Initialize lists to store data
product_brand = []
#product_description = []
price = []
discount = []

# Loop to scrape data from multiple pages (stop when data for 100 sunglasses is collected)

while len(product_brand) < 100:
    # Wait for the page to load
    driver.implicitly_wait(5)
    
    

    # Scrape data from the current page
    product_containers = driver.find_elements(By.CLASS_NAME, "_2B099V")

    for container in product_containers:
        product_name = container.find_element(By.XPATH, './/div[@class="_2WkVRV"]').text
        #product_desc = container.find_element(By.XPATH, '.div[@class=" 
        product_price = container.find_element(By.XPATH, './/div[@class="_30jeq3"]').text
        product_disc = container.find_element(By.XPATH, './/div[@class="_3Ay6Sb"]').text


        product_brand.append(product_name)
        #product_description.append(product_desc)
        price.append(product_price)
        discount.append(product_disc)

        # Check if we have collected data for 100 sunglasses
        if len(product_brand) >= 100:
            break

    # Go to the next page if available
    try:
        next_button = driver.find_element(By.XPATH, "//a[@class='_1LKTO3'][text()='Next']")
        next_button.click()
    except:
        break  # Break the loop if there is no "Next" button

# Create a DataFrame
df = pd.DataFrame({'Brand':product_brand,'Price':price,'Discount':discount})

    

# Display the DataFrame
print(df)


```

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import re
import time

driver=webdriver.Chrome()
driver.get("https://www.flipkart.com/")

shoes_name = driver.find_element(By.CLASS_NAME, "Pke_EE")
shoes_name.send_keys("sneakers")
shoes_name.submit()

sneakers_brand = []
sneakers_description = []
sneakers_prices = []

start = 0
end = 4
for page in range(start,end):
    brand_name = driver.find_elements(By.XPATH, '//div[@class="_2WkVRV"]')

    for i in brand_name:
        brand= i.text
        sneakers_brand.append(brand)
        
    description = driver.find_elements(By.XPATH, '//div[@class="_2B099V"]/a[1]')

    for i in description:
        desc= i.text
        sneakers_description.append(desc)
    
    
    price = driver.find_elements(By.XPATH, '//div[@class="_30jeq3"]')

    for i in price:
        pr= i.text
        sneakers_prices.append(pr)
        
    
    next_button = driver.find_element(By.XPATH, '/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
    next_button.click()
    time.sleep(5)
```

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
import time

driver=webdriver.Chrome()
driver.get("https://www.azquotes.com/")

top_quotes = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]")
top_quotes.click()

quotes_title= []
quotes_author= []
type_of_quotes = []


start = 0
end = 11
for page in range(start,end):
    quote_name = driver.find_elements(By.XPATH, '//a[@class="title"]')

    for i in quote_name:
        quote= i.text
        quotes_title.append(quote)
        
    author_name = driver.find_elements(By.XPATH, '//div[@class="author"]')

    for j in author_name:
        author= j.text
        quotes_author.append(author)
    
    
    type_quote = driver.find_elements(By.XPATH, '//div[@class="tags"]')

    for k in type_quote:
        type_q= k.text
        type_of_quotes.append(type_q)
        
        
    next_button = driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]')
    next_button.click()
    time.sleep(5)

    
```

```python
quotes_df = pd.DataFrame({'Quotes':quotes_title,'Author Name':quotes_author,'Type':type_of_quotes})

quotes_df.head(100)
```

```python
import selenium
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
import time

driver=webdriver.Chrome()
driver.get("https://www.jagranjosh.com/")

gk_explore = driver.find_element(By.XPATH, './/div[@class="Home_OtherCat__kBxvs container"]/ul[1]/li[1]/span/a[1]')
gk_explore.click()


prime_ministers_link = driver.find_element(By.XPATH, "//a[text() = 'List of All Prime Ministers of India (1947-2024)']")
href = prime_ministers_link.get_attribute("href")
driver.get(href)

time.sleep(2)

# Step 4: Scrap the mentioned data
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Find the table containing the data
table = soup.find('div',class_='TableData')

# Extract data from the table
data = []
for row in table.find_all('tr')[1:]:
    columns = row.find_all('td')
    name = columns[1].text.strip()
    born_dead = columns[2].text.strip()
    term_of_office = columns[3].text.strip()
    remarks = columns[4].text.strip()
    data.append([name, born_dead, term_of_office, remarks])

# Create DataFrame
df = pd.DataFrame(data, columns=['Name', 'Born-Dead', 'Term of Office', 'Remarks'])

# Display the DataFrame
print(df)



```
